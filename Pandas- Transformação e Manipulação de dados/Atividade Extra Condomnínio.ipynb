```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dados_cond = pd.read_json('dados_locacao_imoveis.json')
dados_cond.head()
```

```python
dados_cond = pd.json_normalize(dados_cond['dados_locacao'])
dados_cond
```

```python
colunas = list(dados_cond.columns)
colunas
```

```python
dados_cond = dados_cond.explode(colunas[1:]).reset_index(drop=True)
dados_cond
```

```python
dados_cond['valor_aluguel'] = dados_cond['valor_aluguel'].apply(lambda x: x.replace('$','').replace(',','.').replace('reais', '').strip()).reset_index(drop=True)
dados_cond
```

```python
dados_cond['valor_aluguel'] = dados_cond['valor_aluguel'].astype(np.float64)
```

```python
dados_cond.info()
```

```python
dados_cond['apartamento'] = dados_cond['apartamento'].str.replace(' \(blocoAP\)','', regex=True)
dados_cond
```
